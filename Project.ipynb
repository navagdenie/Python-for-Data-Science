### Функция подготовки площади 

```python
def prepare_square(df):
    df.loc[df['Square']<15, 'Square']=df['Square']*10
    df.loc[df['Square']>300, 'Square']=df['Square']/10
    df.loc[df['LifeSquare']<9, 'LifeSquare']=df['LifeSquare']*10
    df.loc[df['LifeSquare']>1000, 'LifeSquare']=df['LifeSquare']/100
    df.loc[df['LifeSquare']>300, 'LifeSquare']=df['LifeSquare']/10
    return df
```

```python
data=prepare_square(data)
```

### Создадим процедуру для обучения и проверки результатов

```python
model1=LR()
model2=Ridge()
model3=RF()
```

```python
def predict_prices(data):
    data.drop(['Id'],axis=1,inplace=True)
    data.drop(['Healthcare_1'],axis=1,inplace=True)
    data['DistrictId']=data['DistrictId'].astype(object)
    data=pd.get_dummies(data)
    X=data.drop(['Price'],axis=1)
    y=data['Price']
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=25)
    model1.fit(X_train,y_train)
    model2.fit(X_train,y_train)
    model3.fit(X_train,y_train)
    y_pred_1=model1.predict(X_test)
    y_pred_2=model2.predict(X_test)
    y_pred_3=model3.predict(X_test)
    y_pred_1_mse=MSE(y_test, y_pred_1)
    y_pred_1_r2=r2(y_test, y_pred_1)
    y_pred_2_mse=MSE(y_test, y_pred_2)
    y_pred_2_r2=r2(y_test, y_pred_2)
    y_pred_3_mse=MSE(y_test, y_pred_3)
    y_pred_3_r2=r2(y_test, y_pred_3)
    print('Model 1 - MSE: {}, R2: {},\nModel 2 - MSE: {}, R2: {},\nModel 3 - MSE: {}, R2: {}'.format(y_pred_1_mse, y_pred_1_r2, y_pred_2_mse, y_pred_2_r2, y_pred_3_mse, y_pred_3_r2))
```

### 1 итерация обучения 

```python
predict_prices(data)
```

```python
def prepare_lifesquare(df):
    df.loc[df['Square']<df['LifeSquare'], 'LifeSquare'] = df['Square']
    square_lifesquare = (df['Square']/df['LifeSquare']).median()
    df['LifeSquare'] = df.LifeSquare.fillna(df['Square']/square_lifesquare)
    df.loc[df['LifeSquare']<12, 'LifeSquare']=df['Square']/square_lifesquare
    return df
```

```python
data = prepare_lifesquare(data)
```

```python
def prepare_kitchensquare(df):
    square_kitchensquare = (df['Square']/df['KitchenSquare']).median()
    df.loc[(df['KitchenSquare']!=0)&(df['KitchenSquare']<5), 'KitchenSquare']=df['Square']/square_kitchensquare
    df.loc[((df['Square']/df['KitchenSquare'])<3)&(df['KitchenSquare']>25), 'KitchenSquare']=df['Square']/square_kitchensquare
    return df
```
